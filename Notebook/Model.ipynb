```python
def average_word_length(x):
    y = x.split()
    z = len(y)
    total_length = 0
    for i in y:
        total_length=total_length+len(i)
    if z>0:
        return total_length/z
    return 0
```

# WordCloudbefore data Pre-processing

```python
def gen_freq(text):
    #Will store the list of words
    word_list = []

    #Loop over all the tweets and extract words into word_list
    for tw_words in text.split():
        word_list.extend(tw_words)

    #Create word frequencies using word_list
    word_freq = pd.Series(word_list).value_counts()
    
    return word_freq
```

```python
def word_cloud(word_frequency):
    from wordcloud import WordCloud

    #Generate word cloud
    wc = WordCloud(width=400, height=330, max_words=100, background_color='white').generate_from_frequencies(word_frequency)

    plt.figure(figsize=(12, 8))
    plt.imshow(wc, interpolation='bilinear')
    plt.axis('off')
    plt.show()
```

```python
word_frequency = gen_freq(df.text.str)
word_cloud(word_frequency)
```

# Data Pre-Processing

```python
import re
import nltk
from nltk.corpus import stopwords
from nltk.tokenize import word_tokenize
from nltk.stem import WordNetLemmatizer
lemmatizer = WordNetLemmatizer()
def clean_text(text):    
    #Remove punctuations
    text = re.sub(r'[?!.;:,#-@-]', '', text)
    text = re.sub('[\\\’]', '', text)
    text = re.sub('[^\w\s]', '', text)
    
    #Remove Stop word and Normalization()
    stop_words = set(stopwords.words('english'))
    word_tokens = word_tokenize(text)
    filtered_sentence = []
    for word in word_tokens:
        word = lemmatizer.lemmatize(word)
        if word not in stop_words and len(word)>2:
            filtered_sentence.append(word)
    text = " ".join(filtered_sentence)

    #Convert to lowercase to maintain consistency
    text = text.lower().replace('\n','').replace('\r','').strip()
    text = re.sub(' +', ' ',text)
    
    return text
```

```python
print(clean_text("house dem aide ’ even see my comey ’ the letter jason"))
```

```python
df['text'] = df['text'].apply(clean_text)
df
```

# Train and test Split

```python
x_train,x_test,y_train,y_test = train_test_split(df['text'],df['label'],test_size = 0.3,random_state = 8)
```

```python
print(x_train.shape)
print(x_test.shape)
print(y_train.shape)
print(y_test.shape)
```

# TFIDF vectorization and Model building with Evaluation matrix

```python
tfidf = TfidfVectorizer(encoding='utf-8',
                        ngram_range=(1,2),
                        stop_words=None,
                        min_df=10,
                        max_df=1.0,
                        max_features=300,
                        norm='l2',
                        sublinear_tf=True)
```

```python
feature_train = tfidf.fit_transform(x_train).toarray()
feature_test  = tfidf.transform(x_test).toarray()
```

```python
from sklearn.ensemble import RandomForestClassifier
from sklearn.metrics import accuracy_score, classification_report
model = RandomForestClassifier()
model.fit(feature_train,y_train)
model_prection = model.predict(feature_test)
print('Accuracy', accuracy_score(y_test,model_prection))
print(classification_report(y_test,model_prection))
```

```python
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import accuracy_score, classification_report
from sklearn.model_selection import KFold
from sklearn.model_selection import cross_val_score
cv = KFold(n_splits=10,random_state=1,shuffle=True)

model = LogisticRegression()
scores = cross_val_score(model,feature_train,y_train,scoring = 'accuracy',cv = cv,n_jobs = -1)
print('Accuracy: %3f (%3f)' % (mean(scores)), std(scores))
```
